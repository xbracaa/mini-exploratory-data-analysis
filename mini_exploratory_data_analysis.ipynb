```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=data, orient="h", palette="Set2")
plt.title("Box Plot - Visualization of Numerical Variables")
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
selected_columns = ['Academic Pressure', 'Study Satisfaction', 'Study Hours', 'Financial Stress', 'Gender', 'Age']
correlation_matrix = data[selected_columns].corr()

print("Matriks Korelasi:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap Korelasi")
plt.show()
```

```python
data_depression = data['Depression'].value_counts()
data_depression
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data["Academic Pressure"], bins=5, color="skyblue", kde=True)
plt.title("Distribusi Tekanan Akademik")
plt.xlabel("Tingkat Tekanan Akademik")
plt.ylabel("Jumlah Pelajar")
plt.show()
```

```python
data_depression.plot(kind='bar', color = 'pink')
plt.title('jumlah data pelajar depresi')
plt.xlabel('depresi')
plt.ylabel('jumlah data')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['Academic Pressure'], data['Study Hours'], color='mediumseagreen', alpha=0.7)
plt.title('Academic Pressure vs Study Hours')
plt.xlabel('Tekanan Akademik')
plt.ylabel('Jam Belajar per Hari')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(data_depression, labels=data_depression.index, autopct='%1.1f%%', startangle=90)
plt.title('Rasio Data Setiap Species')
plt.show()
```
